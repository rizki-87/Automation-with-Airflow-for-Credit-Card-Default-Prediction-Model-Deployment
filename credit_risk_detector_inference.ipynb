```python
import pickle
import pandas as pd
import numpy as np
import random
```

```python
# Load All Files

with open('pipeline.pkl', 'rb') as file_1:
  model = pickle.load(file_1)

```

# Dummy Data for Inference

```python
# Pengaturan Seed
seed_value = 10
random.seed(seed_value)
np.random.seed(seed_value)
n = 100

checking_status_options = ['no checking', '<0','0<=X<200','>=200']
checking_status = random.choices(checking_status_options, k=n)

credit_history_options = ['existing paid', 'critical/other existing credit','delayed previously','all paid','no credits/all paid']
credit_history = random.choices(credit_history_options, k=n)

purpose_options = ['radio/tv', 'new car','furniture/equipment','used car','business','education','repairs','domestic appliance','other','retraining']
purpose = random.choices(purpose_options, k=n)

savings_status_options = ['<100', 'no known savings','100<=X<500','500<=X<1000','>=1000']
savings_status = random.choices(savings_status_options, k=n)

employment_options = ['1<=X<4', '>=7','4<=X<7','<1','unemployed']
employment = random.choices(employment_options, k=n)

personal_status_options = ['male single', 'female div/dep/mar','male mar/wid','male div/sep']
personal_status = random.choices(personal_status_options, k=n)

other_parties_options = ['none', 'guarantor','co applicant']
other_parties = random.choices(other_parties_options, k=n)

property_magnitude_options = ['car', 'real estate','life insurance','no known property']
property_magnitude = random.choices(property_magnitude_options, k=n)

other_payment_plans_options = ['none', 'bank','stores']
other_payment_plans = random.choices(other_payment_plans_options, k=n)

housing_options = ['own', 'rent','for free']
housing = random.choices(housing_options, k=n)

job_options = ['skilled', 'unskilled resident','high qualif/self emp/mgmt','unemp/unskilled non res']
job = random.choices(job_options, k=n)

own_telephone_options = ['none', 'yes']
own_telephone = random.choices(own_telephone_options, k=n)

foreign_worker_options = ['yes', 'no']
foreign_worker = random.choices(foreign_worker_options, k=n)

duration = np.random.uniform(4,72,n)
duration = np.round(duration).astype(int)

credit_amount = np.random.uniform(250,18424,n)
credit_amount = np.round(credit_amount).astype(float)

installment_commitment = np.random.uniform(1,4,n)
installment_commitment = np.round(installment_commitment).astype(object)

residence_since = np.random.uniform(1,4,n)
residence_since = np.round(residence_since).astype(object)

age = np.random.uniform(19,75,n)
age = np.round(age).astype(int)

existing_credits = np.random.uniform(1,4,n)
existing_credits = np.round(existing_credits).astype(int)

num_dependents = np.random.uniform(1,2,n)
num_dependents = np.round(num_dependents).astype(object)



# Membuat DataFrame
data_inf = pd.DataFrame({
    'checking_status': checking_status,
    'credit_history': credit_history,
    'purpose' : purpose,
    'savings_status': savings_status,
    'employment': employment,
    'personal_status' : personal_status,
    'other_parties': other_parties,
    'property_magnitude': property_magnitude,
    'other_payment_plans': other_payment_plans,
    'housing': housing,
    'job': job,
    'own_telephone' : own_telephone,
    'foreign_worker': foreign_worker,
    'duration': duration,
    'credit_amount': credit_amount,
    'installment_commitment': installment_commitment,
    'residence_since': residence_since,
    'age': age,
    'existing_credits': existing_credits,
    'num_dependents': num_dependents
})

data_inf
```

# Prediction 

```python
predict = model.predict(data_inf)
predict
```

```python
prediction = pd.DataFrame(predict,columns=['prediction'])
prediction
```

```python
data_final = pd.concat([data_inf,prediction],axis=1)
data_final
```
